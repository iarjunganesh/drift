# DRIFT grounded-chat results (Tier.LIVE / Terra) — 2026-07-18

This is a frozen, display-only record of a completed bounded **Tier.LIVE** grounded-chat capture answered at `gpt-5.6-terra`. It deliberately contains **no executable cells**, environment/configuration values, database endpoint, provider-budget log, or human review notes.

Terra is the chat-over-knowledge tier. It does not draft or publish Insights; it answers questions using only Insights already human-reviewed and published through the review gate. This run read the reviewed store produced by the earlier `gpt-5.6-sol` capture recorded in [`drift_manual_run.sol.results.ipynb`](drift_manual_run.sol.results.ipynb) and wrote nothing to the database.

It is a companion to the `gpt-5.6-luna` and `gpt-5.6-sol` reviewed-capture records, covering the retrieve-and-answer path rather than the drafting path.

## Recorded outcome

- One bounded question was asked per configured primary source (eight sources), through the same retrieval and grounded-answer code paths the live `/chat` endpoint uses.
- Retrieval ranks the reviewed store by vector distance and always surfaces the nearest Insights, so every question — including those for sources with no reviewed Insight — had candidate evidence placed in front of Terra.
- **Seven of eight** questions were answered with a grounded citation; **one** (PyTorch) was declined outright because retrieval surfaced only unrelated Insights.
- Grounded Insight IDs cited across the run: **10, 11, 13, 15, and 16** — exactly the five reviewed Insights, and no others.
- Terra never cited an Insight ID outside the reviewed store and never invented a release fact.
- This is a small, bounded interaction capture — not continuous or comprehensive release monitoring.

## Retrieve-first boundary

The run shows three distinct, honest behaviours at the grounding boundary:

- **Direct grounding.** TensorRT, vLLM, Transformers, JAX, and NCCL each grounded to their own reviewed Insight and cited only that release's primary source.
- **Explicit decline.** For PyTorch, retrieval surfaced only unrelated Insights (16, 11, 13); Terra answered that it had no PyTorch evidence and cited nothing rather than inventing a release.
- **Grounded negative answer.** For Triton and CUTLASS, Terra also answered that no such Insight exists, but grounded that "no" in the specific reviewed Insights it had inspected — so its citations point at genuine reviewed evidence, never a fabricated Triton or CUTLASS release.

In every case the answer stayed inside the reviewed store. This is the property the capture set out to verify: Terra answers retrieve-first and cites only the evidence it actually used.
